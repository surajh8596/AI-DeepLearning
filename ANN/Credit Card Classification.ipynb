### Drop Unwanted columns

```python
df.drop('ID', axis=1, inplace=True)
df.columns
```

### Split into Dependent & Independent features

```python
x=df.drop('default.payment.next.month', axis=1)
y=df['default.payment.next.month']
```

### Split data into Training and Testind Test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print("x train shape", x_train.shape)
print("x test shape", x_test.shape)
print("y train shape", y_train.shape)
print("y test shape", y_test.shape)
```

### Standardize the Features

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

### Create ANN Model

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

### Inplut Layer

```python
model=Sequential()
model.add(Dense(units=20, activation='relu', kernel_initializer='uniform', input_dim=23))
```

### Hidden Layers

```python
model.add(Dense(units=16, activation='relu', kernel_initializer='normal'))
model.add(Dense(units=8, activation='relu', kernel_initializer='normal'))
model.add(Dense(units=4, activation='relu', kernel_initializer='normal'))
```

### Output Layer

```python
model.add(Dense(units=1, activation='sigmoid', kernel_initializer='normal'))
```

### Model Summary

```python
model.summary()
```

### Compile Layer

```python
from keras.optimizers import Adagrad
opt=Adagrad(lr=0.01)
model.compile(optimizer=opt,loss='binary_crossentropy', metrics=['accuracy'])
```

### Fit and Train the model

```python
hist=model.fit(x_train,y_train, epochs=50, batch_size=20)
```

**After each Epoch Accuracy is increasing and Loss in descreasing.**

### Keys

```python
hist.history.keys()
```

### Plot Loss

```python
plt.figure(figsize=(18,4))
plt.plot(hist.history['loss'], label='Loss')
plt.xticks(np.arange(0,51,1))
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend()
plt.title('Loss in the Model', fontsize=18, color='green');
```

**Loss is reducing after each epoch, loss is constant after 30 epochs.**

### Plot Accuracy

```python
plt.figure(figsize=(18,4))
plt.plot(hist.history['accuracy'], label='Accuracy')
plt.xticks(np.arange(0,51,1))
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend()
plt.title('Accuracy of the Model', fontsize=18, color='green');
```

**Accuracy is increasing after each epochs, After 30 epochs accuracy is constant.**

### Prediction

```python
y_pred=model.predict(x_test)
y_pred=np.where(y_pred>0.5, 1, 0)
y_pred
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_pred,y_test))
print('\n Out of {}, {} are missclassified'.format(y_test.shape[0], (y_test!=y_pred.flatten()).sum()))
```

### Classification Report

```python
print(classification_report(y_test, y_pred))
```

**Accuracy is 83%**

**Precision is 85%**

**Recall Score is 94%**
